```python
import os
import fnmatch 
import pandas as pd
import re
```

```python
def extract_api_endpoints(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()

    # Regex to capture API paths in $fetch calls
    pattern = r"\s*await \$fetch.*\s+.+"
    
    endpoints = re.findall(pattern, content)
    return endpoints
```

```python
fe_uri_dict = []
for root, dir, files in os.walk("./server/api"):
        for items in fnmatch.filter(files, "*.ts"):
            fe_uri_dict.append({
                "fe_uri": root+"/"+items,
                "be_endpoint": extract_api_endpoints(root+"/"+items)
            })
            
        print ("")
```

```python
pd_uri_dict = pd.DataFrame(fe_uri_dict)
pd_uri_dict
```
